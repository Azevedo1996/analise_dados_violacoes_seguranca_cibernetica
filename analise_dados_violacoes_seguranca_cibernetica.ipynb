# Análise de Dados de Violações de segurança cibernética



**Contexto**

Esse projeto consiste em explicar inicialmente os dados que serão utilizados, em sequencia realizaremos uma análise exploratória e por fim, realizado uma conclusão do que conseguimos a partir da análise exploratória.

O Conjunto de dados a ser analisado neste projeto consiste em violações de segurança cibernética que ocorreram no périodo de 2007 a 2014 nos Estados Unidos.

```python
import numpy as np # Computacao numerica
import pandas as pd # Processamento de dados
import matplotlib.pyplot as plt # Vizualização
import seaborn as sb # Plotagem
```

```python
df = pd.read_csv('/content/Cyber Security Breaches.csv')
cols = ['Type_of_Breach','Summary','State','Individuals_Affected','Name_of_Covered_Entity'
        ,'Location_of_Breached_Information','year','Date_Posted_or_Updated']
```

Realizando a contagem de valores únicos para as respectivas colunas Ano, Estado e Tipo de Infração que será utilizado na Análise Explorátoria. Também iremos reorganizar a tabela para facilitar na análise.

```python
year_order = df['year'].value_counts().index
state_order = df['State'].value_counts().index
attack_order = df['Type_of_Breach'].value_counts().index
```

```python
df.head(1100)
```

```python
df = df[cols].reset_index(drop=True)
df = df.loc[df['year']>2006]
df = df.loc[df['State'].isin(state_order[:20])]
df = df.loc[df['Type_of_Breach'].isin(attack_order[:6])]
df['Summary'] = df['Summary'].fillna('No Summary')
```

```python
df.head(1100)
```

A tabela demonstra as linhas inicias e as linhas finais do conjunto de dados que foi importado.



*   São um total de 718 linhas e 8 colunas.
*   As respectivas colunas do conjunto são:
*   Type_of_Breach (Tipo de violação)
*   Summary (Resumo da ocorrência)
*   State (Estado)
*   Individuals_Affected (Pessoas afetadas)
*   Name_of_Covered_Entity (Entidade que sofreu o ataque)
*   Location_of_Breached_Information (Local que ocorreu a violação)
*   year (Ano que ocorreu)
*   Date_Posted_or_Updated (Ano que foi anunciado ou atualização)



# Análise Exploratória
Primeiramente começaremos a explorar as váriaveis uma por vez para identificar tendências nas distribuições e valores átipicos.

Em seguida continuaremos a vizualizar Bivariados, para demonstrar a relação entre as váriaveis nas informações.

Por último, técnicas para vizualizar multivariáveis e identificar a relação entre três ou mais váriaveis ao mesmo tempo.

# Univariada
Iremos analisar uma váriavel por vez para desenhar os gráficos para as seguintes perguntas:

*   Qual o ano com maior número de ataques?
*   Qual é o ciberataque mais comum?
*   Qual estado que mais possui ciberataques?

```python
base_color = sb.color_palette()[0]
year_order = df['year'].value_counts().index
state_order = df['State'].value_counts().index
attack_order = df['Type_of_Breach'].value_counts().index
```

Gerando o gráfico com o ano com mais ataque:

```python
plt.figure(figsize=(20,8))
sb.countplot(data=df,y='year',color=base_color,order=year_order)
plt.xticks(rotation=45);
```

**O maior número de ataque foi em 2013.**

Gerando o gráfico com o ciberataques mais comuns:

```python
plt.figure(figsize=(18,8))
sb.countplot(data=df,y='Type_of_Breach',color=base_color,order=attack_order)
plt.xticks(rotation=15);
```

**Roubo é o tipo de ataque mais comum.**

Gerando o gráfico do estado mais ocorreu ciberataques:

```python
plt.figure(figsize=(20,8))
sb.countplot(data=df,x='State',color=base_color,order=state_order)
plt.xticks(rotation=45);
```

**California é o estado que mais reportou casos de ciberataque.**

# Bivariada
A seguir, vamos analisar a relação entre duas váriaveis e desenhar o gráfico a partir de:

*   Qual a relação entre o Ano (Year) e os Indivíduos Afetados (Individuals Affected)?
*   Qual é a relação entre Ano (Year) e os ciberataques?

Analisando primeiramente a relação que o Ano e os Indivíduos Afetados possuem gerando através de um diagrama de dispersão:

```python
plt.figure(figsize=(20,8))
plt.scatter(data = df, y = 'Individuals_Affected', x = 'year', alpha=1/2)
plt.xticks(rotation=15);
```

**Possuem uma relação fraca mas pode tender a possuir uma relação positiva.**

A seguir, analisando a relação entre anos e os ciberataques:

```python
plt.figure(figsize=(20,8))
sb.countplot(data = df, x = 'year', hue = 'Type_of_Breach',color=base_color,order=year_order);
```

```python
ct_counts = df.groupby(['Type_of_Breach', 'year']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts = ct_counts.pivot(index = 'Type_of_Breach', columns = 'year', values = 'count')
sb.heatmap(ct_counts, annot = True, cmap = 'vlag_r', center = 0)
plt.xticks(rotation=15);
```

Podemos observar que o número de ataques subiu com o tempo.


# Multivariado
Por fim, analisaremos se três variáveis possuem relação a partir de:

Os Anos, ciberataques e individuos afetados tem algum relacionamento?

```python
plt.figure(figsize=(20,8))
ax = sb.barplot(data = df, x = 'year', y = 'Individuals_Affected',
                hue = 'Type_of_Breach',palette='viridis',order=year_order)
plt.xticks(rotation=15);
```

Ao analisar o gráfico, podemos observar que os tipos de ataques que vem aumentando durante os anos é Roubo e Hacking/IT Incident


## Conclusão

*Com base nos gráficos gerados atráves das perguntas que foram propostas a respeito ao conjunto de dados, podemos concluir:*

Roubo é o ciberataque mais recorrente;

*   California é o estado que mais possui casos de ciberataques;
*   2013 é o ano que possui ciberataques relatados;
*   2009-2013 é o período que possui maior número de ataques, especialmente Roubo como demonstrado;
*   Individuos afetados por ciberataques está aumentando a cada ano, porém 2014 é uma exceção;
*   Dois tipo de ataque que está aumentando é Roubo e Hacking/IT incident.
